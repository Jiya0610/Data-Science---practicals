```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("JIYA_MOMIN_dataset(loan_data).csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
Q1 = df["CoapplicantIncome"].quantile(0.25)
```

```python
Q3 = df["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
min_range = Q1 - (1.5*IQR)
```

```python
max_range = Q3 + (1.5*IQR)
```

```python
min_range
```

```python
max_range
```

```python
df = df[df["CoapplicantIncome"]<=max_range]
```

```python
df.head()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
fun = FunctionTransformer(func= np.log1p)
```

```python
fun.fit(df[["CoapplicantIncome"]])
```

```python
df["CoapplicantIncome_tr"] = fun.transform(df[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df["CoapplicantIncome_tr"])
plt.show()
```
