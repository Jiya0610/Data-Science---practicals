```python
df = pd.read_csv("JIYA_MOMIN_dataset(loan_data).csv")
```

```python
df
```

```python
# mean
```

```python
me = df["LoanAmount"].mean()
```

```python
md = df["LoanAmount"].median()
```

```python
mo = df["LoanAmount"].mode()
```

```python
# plot
```

```python
sns.histplot(x = "LoanAmount" , data = df, bins = range(0,600,100) )
plt.plot([me for i in range(0,400)],[i for i in range(0,400)],c = "green" ,label = "Mean")
plt.plot([md for i in range(0,400)],[i for i in range(0,400)],c = "orange",label = "Median")
plt.plot([mo for i in range(0,400)],[i for i in range(0,400)],c = "red",label = "Mode")
plt.legend()
plt.show()
```
