```python
t_Before = np.array([50,60,45,65,55,70,40,75,80,65,70,60,50,55,45,75,60,50,65,70])
t_After = np.array([60,70,55,75,65,80,50,85,90,70,75,65,55,60,50,80,65,55,70,75])
```

```python
std_A = np.std(t_After)
std_B = np.std(t_Before)
mean_A = np.mean(t_After)
mean_B = np.mean(t_Before)
```

```python
t_cal = (mean_A - mean_B)/(np.sqrt(((std_A*std_A)/len(t_After)))+(((std_B*std_B)/len(t_Before))))
```
