```python
#Q:A teacher claims that the mean score of students in his class is greater than 82 with a 
# standard deviation of 20. if a sample of 81 student was selected with a mean score of 90.
```

```python
import scipy.stats as st
import numpy as np
```

```python
Zvalue = st.norm.ppf(0.95) # it gives z value
```

```python
print(Zvalue)
```

```python
x = 90
u = 82
pstd = 20
n = 81
```

```python
Zcal = (x-u)/(pstd/np.sqrt(n))
```

```python
print(Zcal)
```

```python
if (Zcal>Zvalue):
    print("H1 is right")
else:
    print("H0 is right")
```
