```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = [np.random.randint(10,100) for i in range(10000)]
```

```python
np.mean(data)
```

```python
data_table = pd.DataFrame({"Data" : data})
```

```python
data_table
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="Data" , data = data_table)
plt.show()
```

```python
samp_mean = []
for N in range(50):
    sample_data = []
    for i in range(500):
        sample_data.append(np.random.choice(data))
    samp_mean.append(np.mean(sample_data))
```

```python
np.mean(samp_mean)
```

```python
sample_M = pd.DataFrame({"sample_mean":samp_mean})
```

```python
plt.figure(figsize=(3,2))
sns.kdeplot(x="sample_mean",data = sample_M)
plt.show()
```
