```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("JIYA_MOMIN_dataset(loan_data).csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.distplot(df["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(df[["CoapplicantIncome"]])
```

```python
df["co_scale"] = minmax.transform(df[["CoapplicantIncome"]])
```

```python
df.head()
```

```python
plt.figure(figsize=(15,6))
plt.title("Before")
plt.subplot(1,2,1)
sns.distplot(df["CoapplicantIncome"])
plt.title("After")
plt.subplot(1,2,2)
sns.distplot(df["co_scale"])
plt.show()
```
