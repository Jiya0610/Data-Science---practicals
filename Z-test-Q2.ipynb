```python
# Q.2 Scenario : imagine you work for an e-commerce company , and your team is responsible
# for analysing customer purchased data.you want to determine wheather a new website design 
# has led to significant increase in the average  purchased amount compared to the old design.

# Data: you have collected data from a random sample of 30 customer who made purchases on the old 
# website design and 30 customer who made purchases on the new website design .you have th 
# samle mean,sample standard deviation,and sample size for both groups
# - old design data = [45.2,42.8,38.9,43.5,41.0,44.6,40.5,42.7,39.8,41.4,44.3,39.7,42.1,40.6,43.0,42.2
#                     41.5,39.6,44.0,43.1,38.7,43.9,42.0,42.8,43.7,41.3,40.9,42.5,41.6]
# - new design data = [48.5,49.1,50.2,47.8,48.7,49.9,48.0,50.5,49.8,49.6,48.2,48.9,49.7,50.3,49.4
#                     50.1,48.6,49.0,50.0,48.4,49.3,49.5,48.8,50.6,50.4,48.1,49.2,50.7,50.8]
# - population std = 2.5
```

```python
# H0 (old website) = old = new (no changes)
# h1 (new website > old website)
```

```python
import scipy.stats as st
import numpy as np
```

```python
old_design_data = np.array( [45.2,42.8,38.9,43.5,41.0,44.6,40.5,42.7,39.8,41.4,44.3,39.7,42.1,40.6,43.0,42.2,
                   41.5,39.6,44.0,43.1,38.7,43.9, 42.0,42.8,43.7,41.3,40.9,42.5,41.6])
```

```python
new_design_data = np.array( [48.5,49.1,50.2,47.8,48.7,49.9,48.0,50.5,49.8,49.6,48.2,48.9,49.7,50.3,49.4,
                    50.1,48.6,49.0,50.0,48.4,49.3,49.5,48.8,50.6,50.4,48.1,49.2,50.7,50.8])
```

```python
mean_new = np.mean(new_design_data)
```

```python
mean_old = np.mean(old_design_data)
```

```python
Zcal = (mean_new - mean_old)/(2.5/np.sqrt(30))
```

```python
print(Zcal)
```

```python
Zvalue = st.norm.ppf(0.95)
```

```python
print(Zvalue)
```

```python
if Zcal > Zvalue :
    print("H1 is rignt")
else:
    print("H0 is right")
```
